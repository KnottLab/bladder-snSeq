```python
import scanpy as sc
import pandas as pd

%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pairs = [
    ('PVR', 'TIGIT'),
    ('PVR', 'CD96'),
    ('PVR', 'CD226'),
    ('CD274', 'PDCD1'),
    ('LGALS9', 'HAVCR2'),
    ('NECTIN2', 'TIGIT'),
    ('NECTIN2', 'CD226'),
    ('CD80', 'CTLA4'),
    ('CD80', 'CD28'),
    ('PDCD1LG2', 'PDCD1'),
    ('CD86', 'CTLA4'),
    ('CD86', 'CD28'),
]
ad = sc.read_h5ad('../data/MIBC_snSeq.h5ad')
sc.pp.normalize_total(ad, target_sum=10000)
sc.pp.log1p(ad)
obs = pd.read_csv('../data/all_cells.csv', index_col=0, header=0)
ad = ad[obs.index].copy()
ad.obs['subtype'] = obs['subtype']
ad.obs
```

```python
import pandas as pd
import numpy as np
senders = [
 'Proliferating_Epithelial',
 'DSG3_Epithelial',
 'CDH12_CDH18_Epithelial',
 'KRT_Epithelial',
 'Intermediate_Epithelial',
]
receiver = 'Late_Activation_CD8T'

ITX = pd.DataFrame(index=[f'{l}_{r}' for l,r in pairs], columns=senders, dtype=float)
for pair in pairs:
    ligand, receptor = pair
    print(ligand, receptor)
    L = pd.DataFrame({ligand: ad.X[:, ad.var_names==ligand].toarray()[:,0]}, 
                      index=ad.obs_names, dtype=float)
    R = pd.DataFrame({receptor: ad.X[:, ad.var_names==receptor].toarray()[:,0]}, 
                      index=ad.obs_names, dtype=float)
    L['subtype'] = ad.obs.SubType_v3
    R['subtype'] = ad.obs.SubType_v3
    L = L.groupby('subtype').mean()
    R = R.groupby('subtype').mean()
    for sender in senders:
        ITX.loc[f'{ligand}_{receptor}', sender] = L.loc[sender, ligand] * R.loc[receiver, receptor]
        
ITX
```

```python
sns.clustermap(ITX, cmap='Reds', standard_scale=0, row_cluster=False, col_cluster=False,
               lw=1, ec='w', figsize=(4,8))
```
