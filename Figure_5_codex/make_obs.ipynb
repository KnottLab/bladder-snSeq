```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
path = "/storage/codex/datasets_v1/bladder_merged_v10.h5ad"
ad = sc.read_h5ad(path)
ad.obs
```

```python
ad
```

```python
legend = pd.read_csv("codex_sample_legend.csv", index_col=1, header=0)
legend.index = [x+'_v5' for x in legend.index]

snseq_sample_d = {k:legend.loc[k,'sampleID'] for k in legend.index}
patient_d = {k:legend.loc[k,'patient'] for k in legend.index}
biopsy_d = {k:legend.loc[k,'biopsy'] for k in legend.index}

legend
```

```python
snseq_sample = [snseq_sample_d[i] for i in ad.obs.sample_id]
patient = [f'Bladder{patient_d[i]}' for i in ad.obs.sample_id]
biopsy = [biopsy_d[i] for i in ad.obs.sample_id]

ad.obs['snSeq_sample'] = snseq_sample
ad.obs['patient'] = patient
ad.obs['biopsy'] = biopsy
```
