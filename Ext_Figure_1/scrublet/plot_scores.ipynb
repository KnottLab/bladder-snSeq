```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rcParams

rcParams['figure.dpi'] = 300
rcParams['figure.figsize'] = (2,2)
rcParams['figure.facecolor'] = (1,1,1,1)
rcParams['svg.fonttype'] = 'none'
import scanpy as sc
```

```python
sc.settings.figdir = 'scrublet_scores'
```

```python
!ls *_all_cells.h5ad
```

```python
ad = sc.read_h5ad("A_all_cells.h5ad")
print(ad.shape)

sc.pp.normalize_total(ad, target_sum=10000)
sc.pp.log1p(ad)
sc.pp.highly_variable_genes(ad, n_top_genes=int(ad.shape[-1]*0.1))

sc.pp.pca(ad, use_highly_variable=True, n_comps=50)
sc.tl.tsne(ad, use_rep='X_pca', n_jobs=-1)
sc.pl.tsne(ad, color='scrublet_score', s=10, title='A', save='A.svg')
```
